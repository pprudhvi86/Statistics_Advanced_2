Q1. What is hypothesis testing in statistics?

Answer:
A method to decide if sample data provides enough evidence to reject a population assumption (null hypothesis).

Q2. What is the null hypothesis, and how does it differ from the alternative hypothesis?

Answer:
Null Hypothesis (H₀): Assumes no effect or difference.
Alternative Hypothesis (H₁): Suggests a significant effect or difference.
H₀ is the default, H₁ challenges it.

Q3. Explain the significance level in hypothesis testing and its role in deciding the outcome of a test.

Answer:
The significance level (α) is the threshold for rejecting H₀ (commonly 0.05).

If p ≤ α → reject H₀.
If p > α → fail to reject H₀.

Q4. What are Type I and Type II errors? Give examples of each.

Answer:
Type I (α): Rejecting H₀ when it’s true (false positive).
example: A medical test indicates a person has a disease when they actually don’t.
Type II (β): Not rejecting H₀ when it’s false (false negative).
Example: A medical test fails to detect a disease when the person actually has it.

Q5. What is the difference between a Z-test and a T-test? Explain when to use
each.

Answer :
Z-test: Used when population variance (σ²) is known and sample size is large (n > 30).
T-test: Used when population variance is unknown and sample size is small (n ≤ 30).
Summary:

Z-test → Large samples, known variance.
T-test → Small samples, unknown variance.

```python
#Q6. Write a Python program to generate a binomial distribution with n=10 and p=0.5, then plot its histogram.
import numpy as np
import matplotlib.pyplot as plt

# Parameters
n = 10
p = 0.5
size = 1000  # number of samples

# Generate binomial distribution
data = np.random.binomial(n, p, size)

# Plot histogram
plt.hist(data, bins=range(n+2), edgecolor="black", alpha=0.7)
plt.title("Binomial Distribution (n=10, p=0.5)")
plt.xlabel("Number of Successes")
plt.ylabel("Frequency")
plt.show()
```

```python
#Q7. Implement hypothesis testing using Z-statistics for a sample dataset in Python. Show the Python code and interpret the results.
# sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

import numpy as np
from scipy import stats

# Sample data
sample_data = [49.1, 50.2, 51.0, 48.7, 50.5, 49.8, 50.3, 50.7, 50.2, 49.6,
               50.1, 49.9, 50.8, 50.4, 48.9, 50.6, 50.0, 49.7, 50.2, 49.5,
               50.1, 50.3, 50.4, 50.5, 50.0, 50.7, 49.3, 49.8, 50.2, 50.9,
               50.3, 50.4, 50.0, 49.7, 50.5, 49.9]

# Hypothesis: mean = 50
mu_0 = 50
sample_mean = np.mean(sample_data)
sample_std = np.std(sample_data, ddof=1)
n = len(sample_data)

# Z statistic
z_stat = (sample_mean - mu_0) / (sample_std / np.sqrt(n))

# p-value (two-tailed test)
p_value = 2 * (1 - stats.norm.cdf(abs(z_stat)))

print("Sample Mean:", sample_mean)
print("Z-statistic:", z_stat)
print("p-value:", p_value)

# Decision at 5% significance
alpha = 0.05
if p_value < alpha:
    print("Reject Null Hypothesis (Significant Difference)")
else:
    print("Fail to Reject Null Hypothesis (No Significant Difference)")
```

```python
#Q8. Write a Python script to simulate data from a normal distribution and calculate the 95% confidence interval for its mean. Plot the data using Matplotlib

import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Simulate normal data
data = np.random.normal(loc=50, scale=5, size=1000)

# Mean and standard error
mean = np.mean(data)
sem = stats.sem(data)

# 95% Confidence Interval
ci = stats.t.interval(0.95, len(data)-1, loc=mean, scale=sem)

print("Mean:", mean)
print("95% Confidence Interval:", ci)

# Plot histogram
plt.hist(data, bins=30, edgecolor="black", alpha=0.7)
plt.axvline(ci[0], color='red', linestyle='--', label="Lower CI")
plt.axvline(ci[1], color='red', linestyle='--', label="Upper CI")
plt.axvline(mean, color='blue', linestyle='-', label="Mean")
plt.title("Normal Distribution with 95% CI")
plt.legend()
plt.show()
```

```python
#Q9. Write a Python function to calculate the Z-scores from a dataset and visualize the standardized data using a histogram. Explain what the Z-scores represent in terms of standard deviations from the mean.

import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import zscore

# Generate sample data
data = np.random.normal(loc=50, scale=10, size=100)

# Calculate Z-scores
z_scores = zscore(data)

# Plot histogram of standardized data
plt.hist(z_scores, bins=20, edgecolor="black", alpha=0.7)
plt.title("Histogram of Z-scores")
plt.xlabel("Z-score")
plt.ylabel("Frequency")
plt.show()

# Explanation
print("Z-scores represent how many standard deviations a data point is away from the mean.")
```
